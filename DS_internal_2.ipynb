47.Write a python program to implement stack using linked list.

```python
#stack using linked list
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Stack:
    def __init__(self):
        self.top = None

    def is_empty(self):
        return self.top is None

    def push(self, data):
        new_node = Node(data)
        new_node.next = self.top
        self.top = new_node

    def pop(self):
        if self.is_empty():
            return None
        data = self.top.data
        self.top = self.top.next
        return data

    def peek(self):
        if self.is_empty():
            return None
        return self.top.data

    def display(self):
        current = self.top
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")

# Driver code
stack = Stack()

stack.push(10)
stack.push(20)
stack.push(30)
print("Stack after pushes:")
stack.display()

print("Top element (peek):", stack.peek())

print("Popped element:", stack.pop())
print("Stack after one pop:")
stack.display()

print("Is stack empty?", stack.is_empty())
```

51.Write a python program to separate even and odd numbers in a linked list.


```python
#separate even and odd in LL
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

    def separate_even_odd(self):
        if self.head is None:
            return

        even_head = even_tail = None
        odd_head = odd_tail = None
        current = self.head

        while current:
            if current.data % 2 == 0:
                if even_head is None:
                    even_head = even_tail = current
                else:
                    even_tail.next = current
                    even_tail = even_tail.next
            else:
                if odd_head is None:
                    odd_head = odd_tail = current
                else:
                    odd_tail.next = current
                    odd_tail = odd_tail.next
            current = current.next

        #attach odd to tail of even
        if even_tail:
            even_tail.next = odd_head
        if odd_tail:
            odd_tail.next = None

        if even_head:
            self.head = even_head
        else:
            #if no even num in LL then start with odd
            self.head = odd_head

    def display(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")

# Driver code
linked_list = LinkedList()

linked_list.insert(1)
linked_list.insert(2)
linked_list.insert(3)
linked_list.insert(4)
linked_list.insert(5)
linked_list.insert(6)

print("Original linked list:")
linked_list.display()

linked_list.separate_even_odd()

print("Linked list after separating even and odd numbers:")
linked_list.display()
```

52.Write a python program to rotate a linked list.

```python
#rotate LL
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

    def rotate(self, k):
        if self.head is None or k == 0:
            return

        current = self.head
        length = 1
        while current.next:
            current = current.next
            length += 1

        current.next = self.head  # Make the list circular

        k = k % length
        if k == 0:
            return

        steps_to_new_head = length - k
        new_tail = self.head

        for i in range(steps_to_new_head - 1):
            new_tail = new_tail.next

        self.head = new_tail.next
        new_tail.next = None

    def display(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")

# Driver code
linked_list = LinkedList()

# insert elements to the linked list
linked_list.insert(1)
linked_list.insert(2)
linked_list.insert(3)
linked_list.insert(4)
linked_list.insert(5)

print("Original linked list:")
linked_list.display()

# Rotate the list by k nodes
k = 2
linked_list.rotate(k)

print(f"Linked list after rotating by {k} positions:")
linked_list.display()
```

```python
#Not there
class BTreeNode:
    def __init__(self, t, leaf=False):
        self.t = t                # Minimum degree
        self.leaf = leaf          # True if leaf node, else False
        self.keys = []            # List of keys in node
        self.children = []        # List of children (BTreeNode objects)

class BTree:
    def __init__(self, t):
        self.t = t                # Minimum degree of B-tree
        self.root = BTreeNode(t, True)  # Initialize root as a leaf node

    def insert(self, key):
        root = self.root
        if len(root.keys) == 2 * self.t - 1:  # If root is full, split it
            new_node = BTreeNode(self.t, False)
            new_node.children.append(self.root)
            self.split_child(new_node, 0)
            self.root = new_node
        self.insert_non_full(self.root, key)

    # Function to insert a key in a non-full node
    def insert_non_full(self, node, key):
        i = len(node.keys) - 1
        if node.leaf:
            # Find the position to insert the key
            while i >= 0 and key < node.keys[i]:
                i -= 1
            node.keys.insert(i + 1, key)
        else:
            # Find the child to recurse down
            while i >= 0 and key < node.keys[i]:
                i -= 1
            i += 1
            child = node.children[i]
            if len(child.keys) == 2 * self.t - 1:  # If child is full, split it
                self.split_child(node, i)
                if key > node.keys[i]:
                    i += 1
            self.insert_non_full(node.children[i], key)

    # Function to split the child of a node
    def split_child(self, parent, i):
        t = self.t
        node = parent.children[i]
        new_node = BTreeNode(t, node.leaf)

        parent.keys.insert(i, node.keys[t - 1])
        parent.children.insert(i + 1, new_node)

        # Move the second half of the node's keys and children to the new node
        new_node.keys = node.keys[t: ]
        node.keys = node.keys[: t - 1]

        if not node.leaf:
            new_node.children = node.children[t: ]
            node.children = node.children[: t]

    # Function to delete a key from the B-tree
    def delete(self, key):
        self.delete_from_node(self.root, key)
        if len(self.root.keys) == 0:
            if len(self.root.children) > 0:
                self.root = self.root.children[0]

    # Function to delete a key from a node
    def delete_from_node(self, node, key):
        t = self.t
        i = 0
        while i < len(node.keys) and key > node.keys[i]:
            i += 1

        if i < len(node.keys) and node.keys[i] == key:  # If key is found
            if node.leaf:
                node.keys.pop(i)  # Remove key from leaf node
            else:
                self.delete_internal(node, i)
        else:
            if node.leaf:
                return  # Key not found in leaf node
            child = node.children[i]
            if len(child.keys) < t:  # If child has less than t keys, balance it
                self.balance_child(node, i)
            self.delete_from_node(node.children[i], key)

    # Function to delete a key when it's an internal node
    def delete_internal(self, node, i):
        t = self.t
        key = node.keys[i]

        if len(node.children[i].keys) >= t:  # If the left child has enough keys
            pred = node.children[i]
            while not pred.leaf:
                pred = pred.children[len(pred.children) - 1]
            pred_key = pred.keys[-1]
            node.keys[i] = pred_key
            self.delete_from_node(node.children[i], pred_key)
        elif len(node.children[i + 1].keys) >= t:  # If the right child has enough keys
            succ = node.children[i + 1]
            while not succ.leaf:
                succ = succ.children[0]
            succ_key = succ.keys[0]
            node.keys[i] = succ_key
            self.delete_from_node(node.children[i + 1], succ_key)
        else:  # If both children have less than t keys, merge them
            self.merge_children(node, i)
            self.delete_from_node(node.children[i], key)

    # Function to balance a child that has less than t keys
    def balance_child(self, node, i):
        if i > 0 and len(node.children[i - 1].keys) >= self.t:  # Borrow from left sibling
            self.borrow_from_left(node, i)
        elif i < len(node.children) - 1 and len(node.children[i + 1].keys) >= self.t:  # Borrow from right sibling
            self.borrow_from_right(node, i)
        else:  # Merge the child
            self.merge_children(node, i)

    # Function to borrow a key from the left sibling
    def borrow_from_left(self, node, i):
        child = node.children[i]
        sibling = node.children[i - 1]

        child.keys.insert(0, node.keys[i - 1])
        node.keys[i - 1] = sibling.keys.pop()

        if not sibling.leaf:
            child.children.insert(0, sibling.children.pop())

    # Function to borrow a key from the right sibling
    def borrow_from_right(self, node, i):
        child = node.children[i]
        sibling = node.children[i + 1]

        child.keys.append(node.keys[i])
        node.keys[i] = sibling.keys.pop(0)

        if not sibling.leaf:
            child.children.append(sibling.children.pop(0))

    # Function to merge two children
    def merge_children(self, node, i):
        child = node.children[i]
        sibling = node.children[i + 1]

        child.keys.append(node.keys.pop(i))
        child.keys.extend(sibling.keys)

        if not child.leaf:
            child.children.extend(sibling.children)

        node.children.pop(i + 1)

    # In-order traversal of the B-tree
    def inorder_traversal(self, node):
        if node is not None:
            for i in range(len(node.keys)):
                if not node.leaf:
                    self.inorder_traversal(node.children[i])
                print(node.keys[i], end=" ")
            if not node.leaf:
                self.inorder_traversal(node.children[len(node.keys)])

# Driver code
btree = BTree(3)  # B-tree with minimum degree 3 (t = 3)

# Inserting keys
keys = [10, 20, 5, 6, 12, 30, 7, 17]
for key in keys:
    btree.insert(key)

print("In-order traversal of the B-tree after insertion:")
btree.inorder_traversal(btree.root)
print()

# Deleting keys
btree.delete(6)
btree.delete(12)

print("In-order traversal of the B-tree after deletion:")
btree.inorder_traversal(btree.root)
print()
```

69.Write a python program to implement priority queue without heap module.

```python
#mam code
class PriorityQueue:
    def __init__(self):
        self.heap = []

    def parent(self, i):
        return (i - 1) // 2

    def left_child(self, i):
        return 2 * i + 1

    def right_child(self, i):
        return 2 * i + 2

    # Function to insert a new element into the priority queue
    def push(self, value):
        self.heap.append(value)
        self._heapify_up(len(self.heap) - 1)  # Restore the heap property

    # Function to remove and return the element with the highest priority (smallest value)
    def pop(self):
        if len(self.heap) == 0:
            return None
        if len(self.heap) == 1:
            return self.heap.pop()  # If only one element is present

        # Swap the first and last element, remove the last element
        root = self.heap[0]
        self.heap[0] = self.heap.pop()
        self._heapify_down(0)  # Restore the heap property
        return root

    # Function to restore the heap property after insertion
    def _heapify_up(self, index):
        while index != 0 and self.heap[self.parent(index)] > self.heap[index]:
            # Swap if the parent is larger than the current node
            self.heap[self.parent(index)], self.heap[index] = self.heap[index], self.heap[self.parent(index)]
            index = self.parent(index)

    # Function to restore the heap property after deletion
    def _heapify_down(self, index):
        smallest = index
        left = self.left_child(index)
        right = self.right_child(index)

        # Check if left child exists and is smaller than the current node
        if left < len(self.heap) and self.heap[left] < self.heap[smallest]:
            smallest = left

        # Check if right child exists and is smaller than the current smallest node
        if right < len(self.heap) and self.heap[right] < self.heap[smallest]:
            smallest = right

        # If the smallest is not the current node, swap and continue heapifying down
        if smallest != index:
            self.heap[index], self.heap[smallest] = self.heap[smallest], self.heap[index]
            self._heapify_down(smallest)

    # Function to peek at the element with the highest priority (smallest value)
    def peek(self):
        if len(self.heap) == 0:
            return None
        return self.heap[0]

    def is_empty(self):
        return len(self.heap) == 0

    def size(self):
        return len(self.heap)

pq = PriorityQueue()
pq.push(3)
pq.push(5)
pq.push(1)
pq.push(4)

print("Top element:", pq.peek())
print("Popped:", pq.pop())
print("Top element after pop:", pq.peek())
pq.push(2)
print("Popped:", pq.pop())
print("Top element after pop:", pq.peek())
```
