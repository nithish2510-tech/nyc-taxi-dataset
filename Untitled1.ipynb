```python
df['pickup_hour'] = df['pickup_datetime'].dt.hour
df.groupby('pickup_hour')['trip_duration_min'].mean().plot(kind='line', marker='o')
plt.title('Average Trip Duration per Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Avg Duration (min)')
plt.grid(True)
plt.show()
```

```python
df['pickup_day'] = df['pickup_datetime'].dt.day_name()
df['pickup_day'].value_counts().reindex([
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
]).plot(kind='bar', color='coral')
plt.title('Trips by Day of Week')
plt.xlabel('Day')
plt.ylabel('Number of Trips')
plt.grid(axis='y')
plt.show()
```

```python
sns.swarmplot(x='passenger_count', y='trip_duration_min', data=df.sample(500), size=3)
plt.title('Trip Duration by Passenger Count')
plt.xlabel('Passenger Count')
plt.ylabel('Duration (min)')
plt.show()
```

```python
sns.ecdfplot(data=df, x='trip_duration_min')
plt.title('ECDF of Trip Duration')
plt.xlabel('Duration (min)')
plt.ylabel('Proportion')
plt.grid(True)
plt.show()
```

```python
df_sorted = df.sort_values(by='pickup_datetime')
df_sorted['cumulative'] = range(1, len(df_sorted)+1)
plt.step(df_sorted['pickup_datetime'], df_sorted['cumulative'], where='post')
plt.title('Cumulative Trips Step Plot')
plt.xlabel('Time')
plt.ylabel('Cumulative Trips')
plt.show()
```

```python
plt.hist2d(df['pickup_longitude'], df['pickup_latitude'], bins=[100, 100], cmap='plasma')
plt.colorbar()
plt.title('2D Histogram of Pickup Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
df['trip_type'] = np.where(df['trip_duration_min'] < 10, 'Short', 'Long')
sizes = df['trip_type'].value_counts()
plt.pie(sizes, labels=sizes.index, autopct='%1.1f%%', wedgeprops=dict(width=0.4))
plt.title('Donut Chart: Trip Type Distribution')
plt.show()
```

```python
df['pickup_date'] = df['pickup_datetime'].dt.date
trip_counts = df.groupby('pickup_date').size()

plt.figure(figsize=(14, 4))
trip_counts.plot(kind='bar', color='mediumseagreen')
plt.title('Trips per Day')
plt.xlabel('Date')
plt.ylabel('Trip Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['hour'] = df['pickup_datetime'].dt.hour
df['day'] = df['pickup_datetime'].dt.day_name()
pivot = df.groupby(['day', 'hour']).size().unstack(fill_value=0)

pivot.T.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='tab20')
plt.title('Hourly Trips Grouped by Day of Week')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Trips')
plt.legend(title='Day of Week')
plt.show()
```

```python
sample = df.sample(300)
u = sample['dropoff_longitude'] - sample['pickup_longitude']
v = sample['dropoff_latitude'] - sample['pickup_latitude']

plt.figure(figsize=(10, 6))
plt.quiver(sample['pickup_longitude'], sample['pickup_latitude'], u, v,
           angles='xy', scale_units='xy', scale=1, width=0.002, alpha=0.5)
plt.title('Quiver Plot of Taxi Routes (sampled)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

```python
df['hour'] = df['pickup_datetime'].dt.hour
pivot = df.groupby(['hour', 'passenger_count']).size().unstack().fillna(0)
pivot.plot(kind='area', stacked=True, colormap='Accent')
plt.title('Trips per Hour by Passenger Count')
plt.xlabel('Hour')
plt.ylabel('Number of Trips')
plt.show()
```

```python
grouped = df.groupby('passenger_count')['distance_km'].mean()
plt.plot(grouped.values, 'o')
plt.xticks(ticks=range(len(grouped)), labels=grouped.index)
plt.title('Dot Plot: Avg Distance by Passenger Count')
plt.xlabel('Passenger Count')
plt.ylabel('Average Distance (km)')
plt.grid(True)
plt.show()
```
